Task 1: Data Exploration

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("/myntra.csv")
data.head()
```

```python
# Summary statistics
summary_stats = data[['Cost Price', 'Selling Price', 'Expected Demand']].describe()
summary_stats
```

```python
# Visualizations
plt.figure(figsize=(15, 5))

# Histograms
plt.subplot(1, 3, 1)
sns.histplot(data['Cost Price'], bins=20, kde=True)
plt.title('Cost Price Distribution')

plt.subplot(1, 3, 2)
sns.histplot(data['Selling Price'], bins=20, kde=True)
plt.title('Selling Price Distribution')

plt.subplot(1, 3, 3)
sns.histplot(data['Expected Demand'], bins=20, kde=True)
plt.title('Expected Demand Distribution')

plt.tight_layout()
plt.show()
```

1.The histogram of cost price distribution shows that the cost price distribution is skewed to the right, with more products having a lower cost price. The median cost price is around 20, with a significant number of products costing less than 15 and a smaller number of products costing more than 30.

This suggests that the majority of the clothing products in Myntra's inventory are relatively affordable, with most products costing less than the median price. However, there is also a significant number of products that are more expensive, suggesting that there is a demand for higher-priced products in this market.

2.The histogram of selling price distribution shows that the distribution is skewed to the right, with more products having a lower selling price. The median selling price is around 40, with a significant number of products selling for less than 20 and a smaller number of products selling for more than 100.

This suggests that the product is relatively affordable, with most products selling for less than the median price. However, there is also a significant number of products selling for more than the median price, suggesting that there is a demand for higher-priced products in this market.

3.The histogram for Expected demand distribution shows that the expected demand distribution is skewed to the right, with more products having a lower expected demand. The median expected demand is around 500, with a significant number of products having an expected demand of less than 400 and a smaller number of products having an expected demand of more than 700.

This suggests that most products are expected to have a relatively low demand, but there are a few products that are expected to have a high demand. This is important information to consider when making decisions about product selection and inventory management.



Task 2: Profit Analysis

```python
# Calculate Profit
data['Profit'] = (data['Selling Price'] - data['Cost Price']) * data['Expected Demand']
data.head()
```

```python
# Identify products with highest profit potential
high_profit_products = data.nlargest(15,'Profit')
high_profit_products
```

```python
# Visualize top profit products
sns.barplot(x='Profit', y='Product Name', data=high_profit_products)
plt.title('Top 15 Products with Highest Profit Potential')
plt.show()
```

1.The barplot shows that the top 15 products have a wide range of profit potential, from around 5,000 to 20,000. The top 5 products have the highest profit potential, with the coat having the highest profit potential of all.

2.The most profitable products are coats, varsity jackets, blazers, leather jackets, and sweaters.

3.There is a mix of men's, women's, and unisex products in the top 15.

4.The top 15 products represent a variety of different clothing categories, including outerwear, tops, bottoms, and accessories.

5.The profit potential for the top 15 products is significant, with the total profit potential for these products being over 100,000.
